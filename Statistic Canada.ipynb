#### Insertion of nested set in model

```python
from chartofaccounts.models import AccountNode

def create_account_node(data, parent=None, left=1):
    # Create an AccountNode object for the current node
    node = AccountNode.objects.create(
        id=data['id'],
        name=data['name'],
        parent=parent,
        left=left,
        right=left + 1
    )

    # Update the left value for the next node
    left_value = left + 2

    # Recursively create AccountNode objects for children
    for child_data in data.get('children', []):
        left_value = create_account_node(child_data, parent=node, left=left_value)

    # Update the right value for the current node
    node.right = left_value
    node.save()

    return left_value

# Assuming you have a root element with parent=None
root_node = AccountNode.objects.get(parent=None)

# Recompute left and right values for the entire tree
root_node.compute_left_right_values()

# Assuming nested_set is your provided dictionary

# create_account_node(nested_set)

tree = AccountNode.get_tree_with_depth(id=1)  # Force evaluation

for leaf in tree:
    print(f"id::{leaf[0]} parent:{leaf[3]} name:{leaf[2]} {leaf[1]}")


```

### Dictionaire dataframe par produit


```python
product_filter = (df["Products and product groups"]==Products_and_product_groups_list[0])
region_filter = (df["GEO"]==region_list[0])
# date_filter = (df.index.year.isin([2018, 2019, 2020, 2021, 2022, 2023]))
data_dict = {}
for Products_and_product_groups in Products_and_product_groups_list:
    product_filter = (df["Products and product groups"]==Products_and_product_groups)
    data_df = df[product_filter & region_filter][['REF_DATE','VALUE']].copy()
    data_df.set_index("REF_DATE", inplace=True)
    data_df['MonthlyPercentageChange'] = data_df["VALUE"].pct_change()
    annual_inflation = data_df['VALUE'].resample('Y').last().pct_change(fill_method=None) 
    annual_inflation = {k.year:v for k,v in annual_inflation.items()}
    inflation_list = [annual_inflation[k.year] for k, v in data_df.iterrows()]
    data_df["YearInflation"] = inflation_list
    data_df["AnnualisedPercentageChange"] = (data_df['MonthlyPercentageChange'] + 1) ** 12 - 1
    data_dict[Products_and_product_groups] = data_df
    
```
